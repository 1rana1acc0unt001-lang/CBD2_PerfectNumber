```python
import math
def calculate_pMn(p,n):
    return (p**n-1)/(p-1)

def is_pMn_derivedNumber(n,base):
    x=math.log((1+math.sqrt(1+4*base*(base-1)*n))/2,base)
    return x==int(x)

def make_prime_List(n):
    is_prime = [True] * (n + 1)
    is_prime[0], is_prime[1] = False, False
    for i in range(2, int(math.sqrt(n)) + 1):
        if is_prime[i]:
            for j in range(2 * i, n + 1, i):
                is_prime[j] = False
    prime_List=[]
    for i in range(n+1):
        if(is_prime[i]):
            prime_List.append(i)
    return prime_List

def is_prime(n,prime_List):
    if n>prime_List[-1]:
        raise ValueError('prime_List is less than n!!')
    return n in prime_List

def prime_factorize(n,prime_List):
    if n>prime_List[-1]:
        raise ValueError('prime_List is less than n!!')
    factorize_result=[]
    for p in prime_List:
        if p>n:
            break
        if n%p==0:
            count=0
            while n%p==0:
                count+=1
                n=n/p
            factorize_result.append((p,count))
    return  factorize_result

def calculate_tau(factorize_result):
    tau=1
    n=1
    for t in factorize_result:
        p,e=t
        tau*=(p**(e+1)-1)/(p-1)
        n*=p**e
    return  (n,tau-n)

def filter_perfect(n_tau_List,base,ind):
    return list(filter(lambda x:x[0]==(base-1)*x[1]-(base-2)*calculate_pMn(base,ind),n_tau_List))

def make_n_tau_List(n):
    prime_List=make_prime_List(n)
    factorized_List=[prime_factorize(i,prime_List) for i in range(2,prime_List[-1]+1)]
    tau_List=[calculate_tau(t) for t in factorized_List]
    return  tau_List

def estimate_ind(n,tau,base):
    pMn=((base-1)*tau-n)/(base-2)
    return  math.log(1+pMn*(base-1),base)

Num=1000000
n_tau_List=make_n_tau_List(Num)
print('finished')
```
